```python
import pandas as pd
#empty Dataframe
df=pd.DataFrame()
print(df)
```

```python
#dataframe with values
emp=pd.Series(['Parker','John','smith','william'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nSeries to Data frame\n")
print(result)
```

```python
#extracting column from Data Frame
print("\nExtracting one column:\n")
print(result['Emp'])
```

```python
#adding a column
print("\nAdding new column\n")
result['Age']=pd.Series([35,24,40,38])
print (result)
```

```python
print("\nDeleting one column\n")
del result['Age']
print(result)
```

```python
#extracting row feom DataFrame
print("\nExtracting third row:\n")
print(result.loc[2])

```

```python
#slicing
print(result[1:3])
```

```python
d2=pd.DataFrame([['Dale',123],['Mark',143,]],columns=['Emp','ID'])
result=pd.concat([result,d2],ignore_index=True)
print(result)
```
