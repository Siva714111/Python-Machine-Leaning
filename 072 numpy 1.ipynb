```python
#creat a sequence of integer from 0 to 30 in step 5.
import numpy as np
e=np.arange(0,30,5)
print(e)
```

```python
#Reshapping 3x4 array to 2x2 array
arr=np.array([[1,2,3,4],[5,2,4,2],[1,2,0,1]])
newarr= arr.reshape(4,3)
print("/nOriginal array:\n",arr)
print("Reshaped array[4,3]:\n",newarr)
```

```python
#flatten array
flarr=arr.flatten()
print("\nOriginal array:\n",arr)
print("\nFlattened array;\n",flarr)
```

```python
#conversion into types
import numpy as np
arr= np.array([[1,2,3,4],[5,2,4,2],[1,2,0,1]])
print("\n No. of dimensions:\n",arr.ndim)
print("\n Shapeof array:\n",arr.shape)
print("\n Size of array:\n",arr.size)
print("\n Array type:\n",arr.dtype)
newtype=arr.astype('f')
print("\n Converted array element:\n",newtype)
print("\n Converted array type:\n",newtype)      
```
